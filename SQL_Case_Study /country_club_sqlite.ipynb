```python
%%sql
WITH rev AS(SELECT
            f.name as facility_name,
            SUM(CASE WHEN b.memid = 0 THEN b.slots * f.guestcost
                ELSE b.slots * f.membercost END) as total_revenue
            FROM
                Facilities as f
            INNER JOIN
                Bookings as b
            USING (facid)
            GROUP BY f.facid)
SELECT 
    *
FROM
    rev
WHERE 
    total_revenue < 1000
ORDER BY
    total_revenue;
```

```python
%%sql
SELECT
    m.surname || ', ' || m.firstname  as member,
    r.surname || ', ' || r.firstname as recommender
FROM
    Members as m
INNER JOIN
    Members as r
ON 
    m.recommendedby = r.memid
ORDER BY
    m.surname, m.firstname;
```

```python
%%sql
WITH mem_book AS(
    SELECT
        *
    FROM
        Bookings
    WHERE
        memid <> 0)

SELECT
    f.name as facility,
    COUNT(*) as total_usage
FROM
    Facilities as f
INNER JOIN
    mem_book as b
USING 
    (facid)
GROUP BY
    f.name;
```
