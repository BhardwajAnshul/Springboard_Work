Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

```python
increase = pd.DataFrame({'Borough': borough_series})
```

```python
# computing the percentage of increase from each year's average price 
# to the 2020 average price borough-wise and appending it to the 
# increase df
for year in boroughs_yearly.columns[1:-1]:
    price_increase_ratio = (boroughs_yearly[2020].divide(boroughs_yearly[year]) - 1)*100 
    increase['Price Increase from ' + str(year)] = price_increase_ratio
```

```python
# looking at increase
increase
```

```python
# finding 5 largest percentage increases from each year

# initiate empty dictionary
five_largest_df_dict = {}

# iterate each year's top five price increase df
for year in range(1995,2020):
    five_largest_df_dict.update(
        {year : increase.nlargest(5, 'Price Increase from ' + str(year))
                .loc[:,['Borough','Price Increase from ' + str(year)]]
                .reset_index(drop=True)
        })

# concatenate the top five dfs horizontally, multi-indexing the columns with the year as the outer label
five_largest_df = pd.concat(five_largest_df_dict.values(), axis = 1, keys = five_largest_df_dict.keys())

# look at five_largest_df (showing all columns)
pd.set_option('display.max_columns', None)
five_largest_df
```
